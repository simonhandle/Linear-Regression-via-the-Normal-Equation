# Linear Regression via the Normal Equation

```python
# Import the custom module and pathlib + pandas for reading the data
from linear_regression.model import LinearRegression
from pathlib import Path
import pandas as pd
```

```python
path = Path("../data/exam_data.csv")
df = pd.read_csv(path,sep=";")
print(df.shape)
df.head()
```

```python
# Create X_train, y_train and X_test, y_test
df_train = df.sample(n=7) # 7 out of 10 samples will be used for training the model
df_test = df.drop(labels=df_train.index) # The remaining 3 for testing
X_train,y_train = df_train.drop(columns=["id","score"]),df_train["score"]
X_test,y_test = df_test.drop(columns=["id","score"]),df_test["score"]
```

```python
# Define the model, fit our data and predict y from X
model = LinearRegression()
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
# Visualize the predictions of our model (A perfect prediction would result in a straight line of points here)
import matplotlib.pyplot as plt
plt.scatter(y_test,pred)
plt.xlabel("True y")
plt.ylabel("Predicted y")
```

```python
# Evaluate the model (Note: without testing data)
print(f"MSE: {model.mse(y_test,pred):.2f}")
print(f"MAE: {model.mae(y_test,pred):.2f} (On average, the prediction deviates by a total of this)")
```
