```python
%load_ext autoreload
%autoreload 2
from linear_regression import LinearRegression
import pandas as pd
```

```python
df = pd.read_csv("data/exam_data.csv",sep=";")
df.head()
```

```python
X,y = df.drop(columns=["id","score"]),df["score"]
```

```python
model = LinearRegression()
model.fit(X,y)
pred = model.predict(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y,pred)
plt.xlabel("True y")
plt.ylabel("Predicted y")
```

```python
print(f"{model.mse(y,pred):.2f}")
print(f"{model.mae(y,pred):.2f}")
```
